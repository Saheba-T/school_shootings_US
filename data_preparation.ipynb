```python
# Import required packages
import regex as re
import pandas as pd
import numpy as np
```

```python
data_01 = pd.read_csv("school_shootings_US_data_v01.csv")
data_01.head(10)
```

```python
# Take a look at the column data types
data_01.dtypes
```

```python
# Make a copy of the original df
data_02 = data_01.copy()
# Convert the date column from object to datetime
data_02['Date'] = pd.to_datetime(data_01['Date'])
```

```python
# Split the Location column and add new City and State columns to df
data_02[['City','State']] = data_02['Location'].str.split(',', expand = True)
```

```python
# Make a copy of data_02
data_03 = data_02.copy()

# Add new columns for total deaths and injuries by removing [n 1] tag and convert to integer type
data_03['Total_deaths'] = data_03['Deaths'].str.extract(r'(\d+)').astype(int)
data_03['Total_injured'] = data_03['Injuries'].str.extract(r'(\d+)').astype(int)

# Add new column indicating whether culprit or suspect of the shooting incident died
data_03['Culprit_suspect_dead'] = data_02['Deaths'].str.match(r'\d+(\[n 1]){1}')
# Add new column indicating whether culprit or suspect of the shooting incident was injured
data_03['Culprit_suspect_injured'] = data_02['Injuries'].str.match(r'\d+(\[n 1]){1}')

# Drop Deaths and Injuries columns
data_04 = data_03.drop(columns = ['Deaths', 'Injuries'], axis = 1)
```
